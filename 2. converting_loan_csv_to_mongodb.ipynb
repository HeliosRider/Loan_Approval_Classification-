```python
import pandas as pd
from pathlib import Path
```

```python

# Define the file path
file_path = Path("Resources/clean_loan_data.csv")

# Read the CSV file into a DataFrame
loans_df = pd.read_csv(file_path)

# Display the DataFrame
loans_df
```

## Creating the "loan_data_db" database in mongodb 
Load

```python
import pymongo

myclient = pymongo.MongoClient("mongodb://localhost:27017/")

# create the database
loandb = myclient["loan_data_db"]
```

```python
 #Create a collection called "loans":
loans = loandb["loans"]
```

```python
# Convert DataFrame to list of dictionaries
list_of_dicts = loans_df.to_dict(orient='records')
list_of_dicts[0]
```

```python
#load the data into loan_data_db
results=loans.insert_many(list_of_dicts)
```

```python
# list all the databases
print(myclient.list_database_names())
```

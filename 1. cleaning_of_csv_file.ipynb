### Cleaning of the database

```python
# Import of dependencies 
import pandas as pd
from pathlib import Path
```

```python

# Define the file path
file_path = Path("Resources/loan_data_db.csv") # file cleaned in the mongo database 

# Read the CSV file into a DataFrame
loans_df = pd.read_csv(file_path)

# Display the DataFrame
loans_df.head()
```

```python
# Remove _id column
clean_loans_df = loans_df.drop('_id', axis=1)
clean_loans_df.head()
```

```python
# finding the info
clean_loans_df.info()
```

```python
#Dropping all rows that may missing information

df = clean_loans_df.dropna(how='any')
df
```
